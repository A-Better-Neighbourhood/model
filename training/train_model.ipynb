```python
# ============================================================================
# IMPROVED POTHOLE DETECTION TRAINING - MULTIPLE DATASETS + NEGATIVES
# ============================================================================
# Run this in Google Colab with GPU enabled
# ============================================================================

# ============================================================================
# STEP 1: INSTALL PACKAGES
# ============================================================================
print("ğŸ“¦ Installing packages...")
!pip install -q ultralytics roboflow kaggle pyyaml pillow

import os
import shutil
import yaml
import random
from pathlib import Path
from roboflow import Roboflow
from ultralytics import YOLO
from IPython.display import Image as IPImage, display
import glob

print("âœ… Packages installed!\n")

# ============================================================================
# STEP 2: SETUP KAGGLE (For additional datasets)
# ============================================================================
print("=" * 70)
print("ğŸ“¥ SETTING UP KAGGLE API")
print("=" * 70)

# Upload your kaggle.json file when prompted
from google.colab import files
print("\nâš ï¸ Upload your kaggle.json file (from https://www.kaggle.com/settings)")
print("Click 'Choose Files' below:")
uploaded = files.upload()

# Setup Kaggle credentials
!mkdir -p ~/.kaggle
!cp kaggle.json ~/.kaggle/
!chmod 600 ~/.kaggle/kaggle.json

print("âœ… Kaggle API configured!\n")

# ============================================================================
# STEP 3: DOWNLOAD ALL DATASETS
# ============================================================================
print("=" * 70)
print("ğŸ“¥ DOWNLOADING DATASETS")
print("=" * 70)

# Create working directory
os.makedirs('/content/all_datasets', exist_ok=True)
os.chdir('/content/all_datasets')

# -----------------
# Dataset 1: Your Original Roboflow Dataset (Potholes)
# -----------------
print("\n1ï¸âƒ£ Downloading original pothole dataset from Roboflow...")

# Get your API key from: https://app.roboflow.com/settings/api
ROBOFLOW_API_KEY = "YOUR_ROBOFLOW_API_KEY_HERE"
rf = Roboflow(api_key=ROBOFLOW_API_KEY)
ROBOFLOW_WORKSPACE = "YOUR_WORKSPACE_NAME"
ROBOFLOW_PROJECT = "YOUR_PROJECT_NAME"
project = rf.workspace(ROBOFLOW_WORKSPACE).project(ROBOFLOW_PROJECT)

version = project.version(1)
original_dataset = version.download("yolov8", location="/content/all_datasets/original_potholes")
print(f"âœ… Original dataset: {original_dataset.location}")

# -----------------
# Dataset 2: Kaggle Pothole Dataset
# -----------------
print("\n2ï¸âƒ£ Downloading Kaggle pothole dataset...")
!kaggle datasets download -d atulyakumar98/pothole-detection-dataset -p /content/all_datasets/kaggle_potholes
!unzip -q /content/all_datasets/kaggle_potholes/pothole-detection-dataset.zip -d /content/all_datasets/kaggle_potholes
print("âœ… Kaggle pothole dataset downloaded")

# -----------------
# Dataset 3: Annotated Potholes
# -----------------
print("\n3ï¸âƒ£ Downloading annotated potholes dataset...")
!kaggle datasets download -d chitholian/annotated-potholes-dataset -p /content/all_datasets/annotated_potholes
!unzip -q /content/all_datasets/annotated_potholes/annotated-potholes-dataset.zip -d /content/all_datasets/annotated_potholes
print("âœ… Annotated potholes downloaded")

# -----------------
# Dataset 4: Clean Roads (Negative Examples)
# -----------------
print("\n4ï¸âƒ£ Downloading clean roads dataset...")
!kaggle datasets download -d saksham1991/indian-roads-dataset -p /content/all_datasets/clean_roads
!unzip -q /content/all_datasets/clean_roads/indian-roads-dataset.zip -d /content/all_datasets/clean_roads
print("âœ… Clean roads downloaded")

# -----------------
# Dataset 5: Faces (Negative Examples)
# -----------------
print("\n5ï¸âƒ£ Downloading face images (to prevent false positives)...")
!kaggle datasets download -d jessicali9530/lfw-dataset -p /content/all_datasets/faces
!unzip -q /content/all_datasets/faces/lfw-dataset.zip -d /content/all_datasets/faces
print("âœ… Face dataset downloaded")

print("\nğŸ‰ All datasets downloaded!\n")

# ============================================================================
# STEP 4: MERGE AND ORGANIZE DATASETS
# ============================================================================
print("=" * 70)
print("ğŸ”„ ORGANIZING COMBINED DATASET")
print("=" * 70)

# Create merged dataset structure
merged_dataset = '/content/merged_pothole_dataset'
for split in ['train', 'valid', 'test']:
    os.makedirs(f'{merged_dataset}/{split}/images', exist_ok=True)
    os.makedirs(f'{merged_dataset}/{split}/labels', exist_ok=True)

def copy_yolo_dataset(source_path, dest_path, max_images=None):
    """Copy images and labels from source to destination"""
    copied = 0
    for split in ['train', 'valid', 'test']:
        src_img = f"{source_path}/{split}/images"
        src_lbl = f"{source_path}/{split}/labels"
        dst_img = f"{dest_path}/{split}/images"
        dst_lbl = f"{dest_path}/{split}/labels"

        if os.path.exists(src_img):
            images = list(Path(src_img).glob('*.*'))
            if max_images:
                images = random.sample(images, min(len(images), max_images))

            for img_path in images:
                # Copy image
                shutil.copy(img_path, dst_img)

                # Copy label if exists
                label_name = img_path.stem + '.txt'
                label_path = Path(src_lbl) / label_name
                if label_path.exists():
                    shutil.copy(label_path, dst_lbl)

                copied += 1
    return copied

def copy_negative_images(source_path, dest_split, max_images=300):
    """Copy negative examples (no labels)"""
    copied = 0
    dst_img = f"{merged_dataset}/{dest_split}/images"

    images = list(Path(source_path).rglob('*.jpg')) + list(Path(source_path).rglob('*.png'))
    images = random.sample(images, min(len(images), max_images))

    for img_path in images:
        new_name = f"negative_{copied}_{img_path.name}"
        shutil.copy(img_path, f"{dst_img}/{new_name}")
        copied += 1

    return copied

print("\nğŸ“‚ Copying datasets...")

# Copy original pothole dataset (sample 4000 images)
print("   Copying original potholes...")
count1 = copy_yolo_dataset(original_dataset.location, merged_dataset, max_images=4000)
print(f"   âœ… Copied {count1} images from original dataset")

# Copy Kaggle datasets (all images)
print("   Copying Kaggle potholes...")
kaggle_path = '/content/all_datasets/kaggle_potholes'
if os.path.exists(f"{kaggle_path}/train"):
    count2 = copy_yolo_dataset(kaggle_path, merged_dataset)
    print(f"   âœ… Copied {count2} images from Kaggle dataset")

# Copy annotated potholes
print("   Copying annotated potholes...")
annotated_path = '/content/all_datasets/annotated_potholes'
if os.path.exists(f"{annotated_path}/train"):
    count3 = copy_yolo_dataset(annotated_path, merged_dataset)
    print(f"   âœ… Copied {count3} images from annotated dataset")

# Add clean roads (negative examples) - split 70/20/10
print("   Adding clean roads (negative examples)...")
clean_roads_path = '/content/all_datasets/clean_roads'
count4_train = copy_negative_images(clean_roads_path, 'train', max_images=1400)
count4_valid = copy_negative_images(clean_roads_path, 'valid', max_images=400)
count4_test = copy_negative_images(clean_roads_path, 'test', max_images=200)
print(f"   âœ… Added {count4_train + count4_valid + count4_test} clean road images")

# Add faces (negative examples) - split 70/20/10
print("   Adding faces (negative examples)...")
faces_path = '/content/all_datasets/faces'
count5_train = copy_negative_images(faces_path, 'train', max_images=210)
count5_valid = copy_negative_images(faces_path, 'valid', max_images=60)
count5_test = copy_negative_images(faces_path, 'test', max_images=30)
print(f"   âœ… Added {count5_train + count5_valid + count5_test} face images")

# Count final dataset
total_train = len(list(Path(f"{merged_dataset}/train/images").glob('*.*')))
total_valid = len(list(Path(f"{merged_dataset}/valid/images").glob('*.*')))
total_test = len(list(Path(f"{merged_dataset}/test/images").glob('*.*')))

print(f"\nğŸ“Š Final Dataset Composition:")
print(f"   Train: {total_train} images")
print(f"   Valid: {total_valid} images")
print(f"   Test: {total_test} images")
print(f"   Total: {total_train + total_valid + total_test} images")

# ============================================================================
# STEP 5: CREATE DATA.YAML
# ============================================================================
print("\nğŸ“ Creating data.yaml...")

data_yaml = {
    'path': merged_dataset,
    'train': 'train/images',
    'val': 'valid/images',
    'test': 'test/images',
    'nc': 1,
    'names': ['pothole']
}

yaml_path = f'{merged_dataset}/data.yaml'
with open(yaml_path, 'w') as f:
    yaml.dump(data_yaml, f)

print(f"âœ… data.yaml created at: {yaml_path}\n")

# ============================================================================
# STEP 6: TRAIN MODEL WITH IMPROVED SETTINGS
# ============================================================================
print("=" * 70)
print("ğŸ‹ï¸ TRAINING MODEL")
print("=" * 70)

model = YOLO('yolov8n.pt')  # or yolov8s.pt for better accuracy

print("\nâ° Starting training (this will take 2-4 hours)...\n")

results = model.train(
    data=yaml_path,
    epochs=100,                # More epochs for better learning
    imgsz=416,
    batch=16,
    device=0,
    name='pothole_v2_improved',
    patience=20,               # More patience for convergence
    save=True,

    # Advanced augmentation to improve generalization
    augment=True,
    hsv_h=0.015,              # Hue augmentation
    hsv_s=0.7,                # Saturation
    hsv_v=0.4,                # Value/brightness
    degrees=15.0,             # Rotation
    translate=0.1,            # Translation
    scale=0.5,                # Scaling
    flipud=0.0,               # No vertical flip (roads don't flip vertically)
    fliplr=0.5,               # Horizontal flip
    mosaic=1.0,               # Mosaic augmentation
    mixup=0.1,                # Mix images
    copy_paste=0.1,           # Copy-paste augmentation

    # Regularization to prevent overfitting
    dropout=0.0,              # Dropout rate

    # Optimizer settings
    optimizer='auto',         # Auto-select best optimizer
    lr0=0.01,                 # Initial learning rate
    lrf=0.01,                 # Final learning rate
    momentum=0.937,           # SGD momentum
    weight_decay=0.0005,      # Weight decay

    # Validation
    val=True,
    plots=True,               # Generate training plots
    verbose=True
)

print("\nâœ… Training completed!\n")

# ============================================================================
# STEP 7: VALIDATE MODEL
# ============================================================================
print("=" * 70)
print("ğŸ“Š VALIDATING MODEL")
print("=" * 70)

metrics = model.val()

print(f"\nğŸ¯ Performance Metrics:")
print(f"   Precision:     {metrics.box.mp:.3f}")
print(f"   Recall:        {metrics.box.mr:.3f}")
print(f"   mAP@0.5:       {metrics.box.map50:.3f}")
print(f"   mAP@0.5:0.95:  {metrics.box.map:.3f}")

# ============================================================================
# STEP 8: VIEW TRAINING RESULTS
# ============================================================================
print("\n=" * 70)
print("ğŸ“Š TRAINING RESULTS")
print("=" * 70)

model_path = 'runs/detect/pothole_v2_improved/weights/best.pt'
results_path = 'runs/detect/pothole_v2_improved/results.png'
confusion_path = 'runs/detect/pothole_v2_improved/confusion_matrix.png'

if os.path.exists(results_path):
    print("\nğŸ“ˆ Training Curves:")
    display(IPImage(filename=results_path, width=900))

if os.path.exists(confusion_path):
    print("\nğŸ¯ Confusion Matrix:")
    display(IPImage(filename=confusion_path, width=600))

# ============================================================================
# STEP 9: TEST ON DIVERSE IMAGES
# ============================================================================
print("\n=" * 70)
print("ğŸ” TESTING ON DIVERSE IMAGES")
print("=" * 70)

best_model = YOLO(model_path)

# Test on potholes
print("\nâœ… Testing on pothole images...")
results_potholes = best_model.predict(
    source=f'{merged_dataset}/test/images',
    conf=0.25,
    save=True,
    project='runs/detect',
    name='test_potholes'
)

# Test on negative examples (faces)
print("\nâœ… Testing on face images (should NOT detect potholes)...")
face_test_images = list(Path(faces_path).rglob('*.jpg'))[:10]
if face_test_images:
    results_faces = best_model.predict(
        source=[str(img) for img in face_test_images],
        conf=0.25,
        save=True,
        project='runs/detect',
        name='test_faces'
    )

    # Check false positives
    false_positives = sum(1 for r in results_faces if len(r.boxes) > 0)
    print(f"\n   Face images with false detections: {false_positives}/10")
    if false_positives == 0:
        print("   ğŸ‰ Perfect! No false positives on faces!")
    else:
        print(f"   âš ï¸ Still {false_positives} false positives - may need more training")

# Display sample predictions
print("\nğŸ–¼ï¸ Sample Predictions:")
pred_images = sorted(glob.glob('runs/detect/test_potholes/*.jpg'))[:5]
for img_path in pred_images:
    print(f"\nğŸ“¸ {os.path.basename(img_path)}:")
    display(IPImage(filename=img_path, width=700))

# ============================================================================
# STEP 10: SAVE AND DOWNLOAD MODEL
# ============================================================================
print("\n=" * 70)
print("ğŸ’¾ SAVING MODEL")
print("=" * 70)

print(f"\nğŸ“ Model saved at: {model_path}")
print(f"   Model size: {os.path.getsize(model_path) / (1024*1024):.2f} MB")

# Download model
from google.colab import files
print("\nğŸ“¥ Downloading model...")
files.download(model_path)
print("âœ… Download started!")

# Also save to Google Drive (optional)
try:
    from google.colab import drive
    drive.mount('/content/drive')

    drive_path = '/content/drive/MyDrive/pothole_models'
    os.makedirs(drive_path, exist_ok=True)
    shutil.copy(model_path, f'{drive_path}/best_v2_improved.pt')
    print(f"âœ… Model also saved to Google Drive: {drive_path}/")
except:
    print("âš ï¸ Could not save to Drive (optional)")

# ============================================================================
# STEP 11: EXPORT MODEL
# ============================================================================
print("\n=" * 70)
print("ğŸ“¦ EXPORTING MODEL")
print("=" * 70)

print("\nğŸ”„ Exporting to ONNX format...")
try:
    onnx_path = best_model.export(format='onnx', imgsz=416)
    print(f"âœ… ONNX model: {onnx_path}")
    files.download(onnx_path)
except Exception as e:
    print(f"âš ï¸ Export failed: {e}")

# ============================================================================
# SUMMARY
# ============================================================================
print("\n" + "=" * 70)
print("ğŸ‰ TRAINING COMPLETE!")
print("=" * 70)

print(f"""
ğŸ“Š Training Summary:
   â€¢ Total images trained: {total_train + total_valid + total_test}
   â€¢ Epochs: 100
   â€¢ Final mAP@0.5: {metrics.box.map50:.3f}
   â€¢ Model: {model_path}
""")
```
